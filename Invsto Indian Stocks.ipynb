```python
import numpy as np
import pandas as pd
import dask.dataframe as dd
import yfinance as yf
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
import matplotlib.dates as mdates

# Step 1: Data Preparation
stocks = ['TCS.NS', 'RELIANCE.NS', 'HDFCBANK.NS', 'INFY.NS', 'ITC.NS']
data = {}

for stock in stocks:
    try:
        df = yf.download(stock, start='2010-01-01', end='2023-02-26')
        data[stock] = df
    except Exception as e:
        print(f"Error downloading data for {stock}: {e}")
        continue

# Use Dask for handling large datasets
data = {stock: dd.from_pandas(df, npartitions=4) for stock, df in data.items()}

# Data Cleaning
for stock, ddf in data.items():
    ddf = ddf.dropna()  # Drop rows with missing values
    z_scores = ddf.map_partitions(lambda df: np.abs(stats.zscore(df.select_dtypes(include=[np.number]))))
    ddf = ddf[(z_scores < 3).all(axis=1)]
    data[stock] = ddf

# Exploratory Data Analysis (EDA) and Feature Engineering
for stock, ddf in data.items():
    df = ddf.compute()
    
    # Print summary statistics
    print(f"Summary statistics for {stock}:")
    print(df.describe())
    
    # Plot stock price over time
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df['Close'])
    plt.title(f'{stock} Stock Price Over Time')
    plt.show()
    
    # Plot ACF and PACF
    plot_acf(df['Close'])
    plt.title(f'Autocorrelation Function for {stock}')
    plt.show()
    
    plot_pacf(df['Close'])
    plt.title(f'Partial Autocorrelation Function for {stock}')
    plt.show()
    
    # Perform feature engineering
    df.loc[:, f'{stock}_lag1'] = df['Close'].shift(1)
    df.loc[:, f'{stock}_lag5'] = df['Close'].shift(5)
    df.loc[:, f'{stock}_lag10'] = df['Close'].shift(10)
    df.loc[:, f'{stock}_roll_mean'] = df['Close'].rolling(window=20).mean()
    df.loc[:, f'{stock}_roll_std'] = df['Close'].rolling(window=20).std()
    df.loc[:, f'{stock}_pct_change'] = (df['Close'] - df['Close'].shift(1)) / df['Close'].shift(1)
    
    print(f"Feature-engineered data for {stock}:")
    print(df.head())
    
    # Replace cleaned data back into Dask DataFrame
    data[stock] = dd.from_pandas(df, npartitions=4)

# Modeling and Evaluation
for stock, ddf in data.items():
    df = ddf.compute()

    # ARIMA Model
    model = ARIMA(df['Close'], order=(1, 1, 1))
    model_fit = model.fit()
    arima_forecast = model_fit.forecast(steps=30)
    
    # Plot ARIMA forecast
    plt.figure(figsize=(12, 6))
    plt.plot(df['Close'][-30:], label='Actual Prices')
    plt.plot(arima_forecast, label='ARIMA Forecasted Prices')
    plt.title(f'ARIMA Model for {stock}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gcf().autofmt_xdate()
    plt.legend()
    plt.show()

    # Gradient Boosting Model (XGBoost)
    X = df.drop(['Close'], axis=1)
    y = df['Close']

    model = XGBRegressor(missing=np.nan)
    model.fit(X, y)

    # Hyperparameter tuning with GridSearchCV
    params = {'max_depth': [3, 5, 7], 'learning_rate': [0.1, 0.5, 1]}
    grid_search = GridSearchCV(model, params, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X, y)

    print(f'Best Parameters for XGBoost model on {stock}:', grid_search.best_params_)
    print(f'Best Score for XGBoost model on {stock}:', grid_search.best_score_)

    # Re-train XGBoost model with best parameters
    model = XGBRegressor(**grid_search.best_params_, missing=np.nan)
    model.fit(X, y)

    xgboost_forecast = model.predict(X[-30:])
    
    # Calculate metrics for ARIMA and XGBoost models
    arima_rmse = np.sqrt(mean_squared_error(y_true=df['Close'][-30:], y_pred=arima_forecast))
    arima_mae = mean_absolute_error(y_true=df['Close'][-30:], y_pred=arima_forecast)
    arima_mape = mean_absolute_percentage_error(y_true=df['Close'][-30:], y_pred=arima_forecast)
    
    xgboost_rmse = np.sqrt(mean_squared_error(y_true=y[-30:], y_pred=xgboost_forecast))
    xgboost_mae = mean_absolute_error(y_true=y[-30:], y_pred=xgboost_forecast)
    xgboost_mape = mean_absolute_percentage_error(y_true=y[-30:], y_pred=xgboost_forecast)

    # Print metrics
    print(f'ARIMA Model Metrics for {stock}:')
    print('RMSE:', arima_rmse)
    print('MAE:', arima_mae)
    print('MAPE:', arima_mape)
    
    print(f'XGBoost Model Metrics for {stock}:')
    print('RMSE:', xgboost_rmse)
    print('MAE:', xgboost_mae)
    print('MAPE:', xgboost_mape)
```
